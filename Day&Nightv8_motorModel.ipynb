```python
### 클래스 변경
import os

folder_path = '/content/drive/MyDrive/Capstone/labelled_02.24 1300-1330_filtered/labels'  # 예: 'C:/Users/username/Documents/labels'

files = os.listdir(folder_path)

for filename in files:
    if filename.endswith('.txt'):  # .txt 파일만 처리
        file_path = os.path.join(folder_path, filename)

        with open(file_path, 'r') as file:
            lines = file.readlines()

        updated_lines = []
        for line in lines:
            parts = line.split()  # 공백 기준 분리
            if parts[0] == '0':  # 0인 경우 2로 변경
                parts[0] = '2'
            elif parts[0] == '1':  # 1인 경우 3으로 변경
                parts[0] = '3'
            updated_lines.append(' '.join(parts) + '\n')

        with open(file_path, 'w') as file:
            file.writelines(updated_lines)

        print(f"Updated: {filename}")

print("All files have been updated successfully.")
```

```python
# 이미지와 라벨 파일 분류
import os
import shutil

source_dir = "/content/drive/MyDrive/Capstone/labelled_11.15_1800-2000"

images_dir = os.path.join(source_dir, "images")
labels_dir = os.path.join(source_dir, "labels")

os.makedirs(images_dir, exist_ok=True)
os.makedirs(labels_dir, exist_ok=True)

for filename in os.listdir(source_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        shutil.move(os.path.join(source_dir, filename), os.path.join(images_dir, filename))
    elif filename.endswith('.txt') and filename != 'classes.txt':
        shutil.move(os.path.join(source_dir, filename), os.path.join(labels_dir, filename))

print("이미지와 라벨 파일을 각각 분류하여 이동했습니다.")
```

```python
# Train / Test 데이터셋
import random
import shutil

images_dir = "/content/drive/MyDrive/Capstone/labelled_02.24 1300-1330_filtered/images"
labels_dir = "/content/drive/MyDrive/Capstone/labelled_02.24 1300-1330_filtered/labels"

# Train 및 Val 데이터셋 디렉토리
train_images_dir = "/content/drive/MyDrive/Capstone/data/train2/images"
train_labels_dir = "/content/drive/MyDrive/Capstone/data/train2/labels"
val_images_dir = "/content/drive/MyDrive/Capstone/data/val2/images"
val_labels_dir = "/content/drive/MyDrive/Capstone/data/val2/labels"

os.makedirs(train_images_dir, exist_ok=True)
os.makedirs(train_labels_dir, exist_ok=True)
os.makedirs(val_images_dir, exist_ok=True)
os.makedirs(val_labels_dir, exist_ok=True)

all_image_files = os.listdir(images_dir)

random.shuffle(all_image_files)
split_index = int(0.8 * len(all_image_files))
train_files = all_image_files[:split_index]
val_files = all_image_files[split_index:]

for file in train_files:
    shutil.copy(os.path.join(images_dir, file), os.path.join(train_images_dir, file))

    label_file = file.replace('.jpg', '.txt').replace('.png', '.txt')
    if os.path.exists(os.path.join(labels_dir, label_file)):
        shutil.copy(os.path.join(labels_dir, label_file), os.path.join(train_labels_dir, label_file))

for file in val_files:
    shutil.copy(os.path.join(images_dir, file), os.path.join(val_images_dir, file))

    label_file = file.replace('.jpg', '.txt').replace('.png', '.txt')
    if os.path.exists(os.path.join(labels_dir, label_file)):
        shutil.copy(os.path.join(labels_dir, label_file), os.path.join(val_labels_dir, label_file))

print("Train과 Val 데이터셋으로 파일을 분류하였습니다.")
```
